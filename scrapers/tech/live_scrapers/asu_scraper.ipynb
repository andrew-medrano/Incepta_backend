```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Setup Chrome options
options = webdriver.ChromeOptions()
# Add headers
options.add_argument('--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36')
# Add headless mode if needed
# options.add_argument('--headless')

# Initialize driver with options
driver = webdriver.Chrome(options=options)

try:
    # Base URL for search
    base_url = "https://skysonginnovations.com/?s="
    
    # Navigate to the page
    driver.get(base_url)
    
    # Wait for dynamic content to load (adjust selector and timeout as needed)
    wait = WebDriverWait(driver, 10)
    search_results = wait.until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, "ais-hits--content"))
    )
    
    # Extract and print results
    for result in search_results:
        print(result.text)
        print('-' * 50)

finally:
    # Always close the browser
    driver.quit()
```

```python
def get_titles_links(driver):
    titles = []
    links = []
    
    # Wait for dynamic content to load, just like in the working example
    wait = WebDriverWait(driver, 10)
    elements = wait.until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, "ais-hits--content"))
    )
    
    for element in elements:
        # Find the title and link within each result element
        title = element.find_element(By.TAG_NAME, "a").text
        # You might need to adjust this selector depending on the actual HTML structure
        link_element = element.find_element(By.TAG_NAME, "a")
        link = link_element.get_attribute('href')
        
        titles.append(title)
        links.append(link)
    
    return titles, links

# Initialize driver with options
driver = webdriver.Chrome(options=options)

try:
    # Base URL for search
    base_url = "https://skysonginnovations.com/?s="
    
    # Navigate to the page
    driver.get(base_url)
    
    # Get titles and links
    titles, links = get_titles_links(driver)
    print([i for i in zip(titles, links)])

finally:
    # Always close the browser
    driver.quit()
```
