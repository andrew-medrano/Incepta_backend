```python
import requests
from bs4 import BeautifulSoup
from typing import List, Tuple
```

```python


base_url = 'https://tlo.mit.edu/industry-entrepreneurs/available-technologies'

all_pages = [base_url + f'?page={i}' for i in range(0, 133)]

first_page = all_pages[0]
first_page_soup = BeautifulSoup(requests.get(first_page).text, 'html.parser')

print(first_page_soup.prettify())
```

```python
titles = first_page_soup.find_all('a', class_="arrow-text arrow-text--hide tech-brief-teaser__link")
for title in titles:
    title_text = title.get_text(strip=True)
    link = title['href']

def get_titles_and_links(soup: BeautifulSoup) -> List[Tuple[str, str]]:
    titles = soup.find_all('a', class_="arrow-text arrow-text--hide tech-brief-teaser__link")
    return [(title.get_text(strip=True), title['href']) for title in titles][1:]

titles_and_links = get_titles_and_links(first_page_soup)
print(titles_and_links)
```

```python
results_page_url = 'https://tlo.mit.edu' + titles_and_links[0][1]
print(results_page_url)
results_page_soup = BeautifulSoup(requests.get(results_page_url).text, 'html.parser')
print(results_page_soup.prettify())

# https://tlo.mit.edu/industry-entrepreneurs/available-technologies/drinkable-tough-hydrogels-gastrointestinal
```

```python
intro = results_page_soup.find('div', class_="tech-brief-details__intro").text.strip()
# print(intro)

body = results_page_soup.find('div', class_="tech-brief-body")

# First, let's check what headings exist
headings = body.find_all('h2')
print("All h2 headings found:", [h.text.strip() for h in headings])

# Try finding with case-insensitive and stripped text
technology = body.find('h2', string=lambda text: text and text.strip() == 'Technology').find_next('p').text.strip()
print("\nTechnology:", technology)

problem_addressed = body.find('h2', string=lambda text: text and text.strip() == 'Problem Addressed').find_next('p').text.strip()
print("\nProblem addressed:", problem_addressed)

advantages = body.find('h2', string=lambda text: text and text.strip() == 'Advantages').find_next('ul').get_text(strip=True, separator='. ')
print("\nAdvantages:", advantages)
```
