```python
base_url = 'https://upenn.technologypublisher.com/?Test_Inteum_Tech_Publisher_PCI%5Bpage%5D=1'
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time

# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument('--headless=new')  # Use new headless mode
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Initialize the driver
driver = webdriver.Chrome(options=chrome_options)

try:
    # Load the page
    driver.get(base_url)
    time.sleep(2)  # Wait for dynamic content to load
    
    # # Scroll down a few times to load more content
    # body = driver.find_element('tag name', 'body')
    # for _ in range(5):
    #     body.send_keys(Keys.PAGE_DOWN)
    #     time.sleep(0.5)  # Wait for content to load after each scroll
    
    # Get the page source and parse with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    print("Successfully retrieved and parsed page")
    
    print(soup.prettify())

finally:
    # Make sure to close the browser
    driver.quit()
```
